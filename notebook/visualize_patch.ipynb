```python
# Python Notebook: Visualize painted maps and compute power spectrum using pixell
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
from pixell import enmap, enplot
```

```python
# ------------------ parameters ------------------
map_dir = Path("../map_patches_demo")  # folder where Julia saved the maps
idx_range = range(0, 2)                 # adjust to the indices you want to load
```

```python
# ------------------ function to load FITS map ------------------
def load_map(idx):
    fname = map_dir / f"y_map_10deg_fast_{idx}.fits"
    if not fname.exists():
        raise FileNotFoundError(f"Map file {fname} not found")
    # Convert Path object to string for enmap.read_map
    return enmap.read_map(str(fname))
```

```python
idx = 0
m = load_map(idx)
# Adjust font size and title for better visibility
plot = enplot.plot(m, colorbar=True, ticks=True, font_size=14)
# Show the plot with the specified title
enplot.show(plot)
```
